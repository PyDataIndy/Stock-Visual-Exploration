![Image of Plotly](https://www.fastly.com/sites/default/files/Plotly.png)

Plotly is perhaps one of the better known visualization libraries supported across many different programming languages .  Imagine the color styling you might appreciate from seaborn, add some JavaScript interactives, and pre-build datasets via cloud APIs and you get a few things like the following...

````python
import plotly.plotly as py
import cufflinks as cf

# Cufflinks is a mid-tier module built between the Plotly visualization library and Pandas DataFrames.
````

````python
end_plotly = datetime.today()
start_plotly = datetime(end_plotly.year - 1, end_plotly.month, end_plotly.day)

df_aapl = web.DataReader("AAPL", 'yahoo', start_plotly, end_plotly)
df_amzn = web.DataReader("AMZN", 'yahoo', start_plotly, end_plotly)
df_fb = web.DataReader("FB", 'yahoo', start_plotly, end_plotly)
df_goog = web.DataReader("GOOG", 'yahoo', start_plotly, end_plotly)
df_ibm = web.DataReader("IBM", 'yahoo', start_plotly, end_plotly)
df_msft = web.DataReader("MSFT", 'yahoo', start_plotly, end_plotly)

df = pd.DataFrame({'Apple': df_aapl.Open, 'Amazon': df_amzn.Open, 'Facebook' : df_fb.Open,
                   'Google': df_goog.Open, 'IBM': df_ibm.Open, 'MSFT': df_msft.Open})

df.tail()
````

````python
# Plotly Cloud API Source and Demo Account Used Below...

py.sign_in('Python-Demo-Account', 'gwt101uhh0')

df.iplot(kind='line', fill=False, yTitle='Open Price', title='Top Tech Stocks',
        filename='cufflinks/stock data', dimensions=(950, 450), world_readable=True)
````
